```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("mushrooms.csv") #https://www.kaggle.com/datasets/uciml/mushroom-classification
```

```python
df
```

```python
df2 = pd.get_dummies(df)
```

```python
df2
```

```python
import seaborn as sns
```

```python
df2 = df2.apply(np.float64)
```

```python
df2
```

```python
sns.heatmap(df2.corr())
```

```python
from sklearn.decomposition import PCA
```

```python
df3 = df2.drop(['class_e', 'class_p'], axis=1)
```

```python
pca = PCA(n_components=2)

#random is for better visualization, because all the data comes from ints
pca1 = pca.fit_transform(df2).T# + np.random.rand(8124, 2).T*0.05-0.025
pca2 = pca.fit_transform(df3).T# + np.random.rand(8124, 2).T*0.05-0.025
```

```python
pca1.shape
```

```python
sns.scatterplot(x = pca1[0], y = pca1[1], hue=df2.class_e)
```

```python
sns.scatterplot(x = pca2[0], y = pca2[1], hue=df2.class_e)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
kmeans1 = KMeans(n_clusters=2).fit(df2)
```

```python
kmeans2 = KMeans(n_clusters=2).fit(df3)
```

```python
kmeans3 = KMeans(n_clusters=23).fit(df2)
```

```python
kmeans4 = KMeans(n_clusters=23).fit(df3)
```

```python
km = kmeans1.fit_predict(df2)
sns.scatterplot(x = pca1[0], y = pca1[1], hue=km)
```

```python
km = kmeans2.fit_predict(df3)
sns.scatterplot(x = pca2[0], y = pca2[1], hue=km)
```

```python
km = kmeans3.fit_predict(df2)
sns.scatterplot(x = pca1[0], y = pca1[1], hue=km)
```

```python
km = kmeans4.fit_predict(df3)
sns.scatterplot(x = pca2[0], y = pca2[1], hue=km)
```

```python
import matplotlib.pyplot as plt
```

```python
inertia = []
for k in range(1,30):
    kmeans = KMeans(n_clusters=k, random_state=1).fit(df2)
    inertia.append(np.sqrt(kmeans.inertia_))

plt.plot(range(1, 30), inertia, marker='s');
plt.xlabel('$k$')
plt.ylabel('Variance')
```

```python
km = KMeans(n_clusters=10).fit(df2).fit_predict(df2)
sns.scatterplot(x = pca1[0], y = pca1[1], hue=km)
```
